```python
import pandas as pd
import spacy, en_core_web_lg
import swifter
```

```python
df = pd.read_csv("finalized_extreme_copy.csv")
```

```python
nlp = spacy.load("en_core_web_lg")
def get_iob(words):
    doc = nlp(words)
    iob = []
    for token in doc:
        iob.append(token.ent_iob_)
    print(iob)
df["words"].swifter.apply(get_iob)
```
