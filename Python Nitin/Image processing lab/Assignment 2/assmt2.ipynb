```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.exposure import match_histograms
```

# Histogram Equalization

```python
# Perform the Histogram Equalization for "image 1"
img = cv2.imread("1.jpg", 0)
eq_img = cv2.equalizeHist(img)
res = np.hstack((img, eq_img))

plt.title("L. (Original Image), R. (Histogram Equalized Image)")
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB), cmap = "gray")
plt.axis("off")

plt.show()
```

# Histogram Matching

```python
# Perform the Histogram Specification / Matching for the image2. Take the histogram of "imagex as the specified histogram.

img = cv2.imread("/content/6.png", 0)
img2 = cv2.imread("/content/7.png", 0)
matched = match_histograms(img, img2)

plt.figure(figsize = (10, 10))

plt.subplot(131)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(132)
plt.title("Reference Image")
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(133)
plt.title("After Histogram Matching")
plt.imshow(matched, cmap = "gray")
plt.axis("off")

plt.show()
```

# Image Filters

```python
# Apply 3 * 3 and 5 * 5 filter on the "image3" and "image4" respectively
img3 = cv2.imread("img3.png", 0)
img4 = cv2.imread("img4.png", 0)

md_filtered3 = cv2.medianBlur(img3, 3)
md_filtered4 = cv2.medianBlur(img4, 5)

figure = plt.figure(figsize = (10, 10))

plt.subplot(221)
plt.title("Original Image 3")
plt.imshow(img3, cmap = "gray")
plt.axis("off")

plt.subplot(222)
plt.title("Median Blur Filter Applied")
plt.imshow(md_filtered3, cmap = "gray")
plt.axis("off")

plt.subplot(223)
plt.title("Original Image 4")
plt.imshow(img4, cmap = "gray")
plt.axis("off")

plt.subplot(224)
plt.title("Median Blur Filter Applied ")
plt.imshow(md_filtered4, cmap = "gray")
plt.axis("off")

plt.show()
```

```python
# Apply 3 * 3 and 5 * 5 mean filter on "image3" & "image4"
img3 = cv2.imread("img3.png", 0)
img4 = cv2.imread("img4.png", 0)

mean_filtered3 = cv2.blur(img3, (3, 3))
mean_filtered4 = cv2.blur(img4, (5, 5))

figure = plt.figure(figsize = (10, 10))

plt.subplot(221)
plt.title("Original Image 3")
plt.imshow(img3, cmap = "gray")
plt.axis("off")

plt.subplot(222)
plt.title("Mean Blur Filter Applied")
plt.imshow(mean_filtered3, cmap = "gray")
plt.axis("off")

plt.subplot(223)
plt.title("Original Image 4")
plt.imshow(img4, cmap = "gray")
plt.axis("off")

plt.subplot(224)
plt.title("Median Blur Filter Applied ")
plt.imshow(mean_filtered4, cmap = "gray")
plt.axis("off")

plt.show()
```

```python
# Apply the following weighted mean filter to "image3" & "image4"
#          [[1, 2, 1],
# (1/16)   [2, 4, 2],
#         [1, 2, 1]]

img3 = cv2.imread("img3.png", 0)
img4 = cv2.imread("img4.png", 0)
kernel = np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]]) / 16

filtered_img3 = cv2.filter2D(img3, -1, kernel)
filtered_img4 = cv2.filter2D(img4, -1, kernel)

figure = plt.figure(figsize = (10, 10))

plt.subplot(221)
plt.title("Original Image 3")
plt.imshow(img3, cmap = "gray")
plt.axis("off")

plt.subplot(222)
plt.title("Mean Blur Filter Applied")
plt.imshow(filtered_img3, cmap = "gray")
plt.axis("off")

plt.subplot(223)
plt.title("Original Image 4")
plt.imshow(img4, cmap = "gray")
plt.axis("off")

plt.subplot(224)
plt.title("Median Blur Filter Applied ")
plt.imshow(filtered_img4, cmap = "gray")
plt.axis("off")

plt.show()
```
