Point - Point transformations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("image.png", cv2.IMREAD_GRAYSCALE)
threshold_value = 128
_, thresholded_image = cv2.threshold(image, threshold_value, 256, cv2.THRESH_BINARY)
histogram_of_image = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.figure(figsize = (20, 5))
plt.subplot(1, 2, 1)
plt.imshow(thresholded_image, "gray")
plt.title("Thresholded image")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.plot(histogram_of_image)
plt.title("Histogram of the original image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequence of Pixels")

plt.show()
```

Histogram Equalization

```python
image = cv2.imread("image.png", cv2.IMREAD_GRAYSCALE)
hist_image = cv2.calcHist([image], [0], None, [256], [0, 256])
h_eq_image = hist_equalized = cv2.equalizeHist(image)
hist_eq_image = cv2.calcHist([h_eq_image], [0], None, [256], [0, 256])

plt.figure(figsize = (20, 10))
plt.subplot(2, 2, 1)
plt.imshow(image, "gray")
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.plot(hist_image)
plt.title("Histogram of the original image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequence of Pixels")

plt.subplot(2, 2, 3)
plt.imshow(h_eq_image, "gray")
plt.title("Histogram Equalized Image")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.plot(hist_eq_image, scalex = 0.5, scaley = 0.5)
plt.title("Histogram of the Hist Equalized image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequence of Pixels")

plt.tight_layout()
plt.show()
```
